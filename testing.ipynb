```python
from PIL import Image
import google.generativeai as genai
import os
```

```python
defect_image = Image.open("defects.jpeg")
```

```python
defect_image
```

```python
gemini_api_key = os.getenv("GOOGLE-API-19jan-key2")
genai.configure(api_key=gemini_api_key)
model = genai.GenerativeModel('gemini-2.5-flash-lite')
```

```python
prompt = 'Identify the structural defect in the image and suggest me the fix to resolve this issue.'
```

```python
response = model.generate_content([prompt,defect_image],generation_config={"temperature":0.7})
```

```python
print(response.text)
```
